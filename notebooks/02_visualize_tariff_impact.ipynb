```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
nifty = pd.read_csv('../data/nifty50_close_prices_2025.csv', parse_dates=['Date'])
tariffs = pd.read_csv('../data/tariff_events_2025.csv', parse_dates=['Date'])

# Calculate average close price (excluding 'Date' column)
nifty['Average_Close'] = nifty.drop(columns=['Date']).mean(axis=1)


# Compute percent change in average close price around each event
window = 5  # days before and after
impact_results = []

for idx, row in tariffs.iterrows():
    event_date = row['Date']
    event_desc = row['Event']

    # Get index of the event date in nifty
    if event_date not in nifty['Date'].values:
        continue  # skip if date doesn't exist in data

    event_idx = nifty.index[nifty['Date'] == event_date][0]

    # Get 5-day window before and after
    pre_window = nifty.iloc[max(event_idx - window, 0):event_idx]['Average_Close'].mean()
    post_window = nifty.iloc[event_idx + 1:event_idx + 1 + window]['Average_Close'].mean()

    # Calculate percent change
    pct_change = ((post_window - pre_window) / pre_window) * 100
    impact_results.append({
        'Date': event_date,
        'Event': event_desc,
        'Pre Avg': round(pre_window, 2),
        'Post Avg': round(post_window, 2),
        '% Change': round(pct_change, 2)
    })

# Convert to DataFrame for display
impact_df = pd.DataFrame(impact_results)
print(impact_df)

# Save event-wise price change data to CSV
impact_df.to_csv("../data/price_changes.csv", index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

nifty['Average_Close'] = nifty.drop(columns=['Date']).mean(axis=1)

# Calculate 10-day rolling volatility (standard deviation)
nifty['Volatility'] = nifty['Average_Close'].rolling(window=10).std()

# Plot
plt.figure(figsize=(14, 7))
plt.plot(nifty['Date'], nifty['Volatility'], label='10-Day Rolling Volatility', color='purple')

# Mark tariff events
for idx, row in tariffs.iterrows():
    event_date = row['Date']
    event_desc = row['Event']
    
    plt.axvline(x=event_date, color='red', linestyle='--', linewidth=1)
    plt.annotate(
        event_desc,
        xy=(event_date, nifty['Volatility'].max()*0.8),
        xytext=(event_date, nifty['Volatility'].max()),
        arrowprops=dict(facecolor='red', arrowstyle='->'),
        fontsize=8,
        rotation=90,
        ha='center'
    )

# Formatting
plt.title('NIFTY 50 - Volatility Around Tariff Events (2025)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volatility (Std Dev of Average Close)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Define sectors and their stocks
sectors = {
    'Tech': ['INFY.NS', 'TCS.NS', 'HCLTECH.NS', 'TECHM.NS', 'WIPRO.NS'],
    'Pharma': ['SUNPHARMA.NS', 'CIPLA.NS', 'DIVISLAB.NS', 'DRREDDY.NS', 'APOLLOHOSP.NS'],
    'Banking & Finance': ['HDFCBANK.NS', 'ICICIBANK.NS', 'AXISBANK.NS', 'KOTAKBANK.NS', 'SBIN.NS',
                          'BAJFINANCE.NS', 'BAJAJFINSV.NS', 'HDFCLIFE.NS', 'ICICIPRULI.NS', 'SBILIFE.NS'],
    'Energy': ['RELIANCE.NS', 'ONGC.NS', 'POWERGRID.NS', 'NTPC.NS', 'COALINDIA.NS', 'BPCL.NS'],
    'Auto': ['TATAMOTORS.NS', 'MARUTI.NS', 'EICHERMOT.NS', 'BAJAJ-AUTO.NS', 'HEROMOTOCO.NS', 'M&M.NS'],
    'Infra': ['LT.NS', 'GRASIM.NS', 'ULTRACEMCO.NS', 'TATASTEEL.NS', 'JSWSTEEL.NS'],
    'FMCG': ['HINDUNILVR.NS', 'ITC.NS', 'BRITANNIA.NS', 'NESTLEIND.NS', 'TATACONSUM.NS'],
    'Others': ['BHARTIARTL.NS', 'ADANIENT.NS', 'ADANIPORTS.NS', 'UPL.NS', 'INDUSINDBK.NS']
}

# Calculate average price per sector
for sector, stocks in sectors.items():
    valid_stocks = [s for s in stocks if s in nifty.columns]  # ignore missing ones
    nifty[sector + '_Avg'] = nifty[valid_stocks].mean(axis=1)

# Plot all sectors
plt.figure(figsize=(16, 9))

for sector in sectors.keys():
    plt.plot(nifty['Date'], nifty[sector + '_Avg'], label=sector)

plt.title('Sector-Wise Average Price Trends in NIFTY 50 (2025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Average Sector Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 📂 Load tariff events data
tariff_events = pd.read_csv('../data/tariff_events_2025.csv', parse_dates=['Date'])

# ✅ Define sector_df from nifty
sector_columns = [s + '_Avg' for s in sectors.keys()]
sector_df = nifty[['Date'] + sector_columns].copy()
sector_df.set_index('Date', inplace=True)

# 🔄 Normalize sector-wise data (divide by first value)
normalized_sector_df = sector_df.copy()
for sector in sector_df.columns:
    normalized_sector_df[sector] = sector_df[sector] / sector_df[sector].iloc[0]

# 📉 Plot normalized sector price trends
plt.figure(figsize=(14, 7))
for sector in normalized_sector_df.columns:
    plt.plot(normalized_sector_df.index, normalized_sector_df[sector], label=sector)

# 📍 Add vertical tariff event lines
for i, row in tariff_events.iterrows():
    plt.axvline(row['Date'], color='red', linestyle='--', linewidth=1)
    plt.text(row['Date'], 1.05, row['Event'], rotation=90, verticalalignment='bottom', fontsize=8)

# 📋 Chart formatting
plt.title("Normalized Sector-Wise Price Trends in NIFTY 50 (2025)")
plt.xlabel("Date")
plt.ylabel("Normalized Price (Base = 1.0)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define sectors (matching column names used in your `nifty` DataFrame)
sector_names = ['Tech_Avg', 'Pharma_Avg', 'Banking & Finance_Avg', 'Energy_Avg',
                'Auto_Avg', 'Infra_Avg', 'FMCG_Avg', 'Others_Avg']

# Step 2: Set Date index for easier slicing
nifty.set_index('Date', inplace=True)
nifty.index = pd.to_datetime(nifty.index)
tariff_events['Date'] = pd.to_datetime(tariff_events['Date'])

# Step 3: Calculate 5-day before and after average returns
impact_matrix = pd.DataFrame(index=sector_names)

for _, row in tariff_events.iterrows():
    event_date = row['Date']
    event_name = row['Event']
    
    before_window = nifty.loc[event_date - pd.Timedelta(days=7): event_date - pd.Timedelta(days=1)]
    after_window = nifty.loc[event_date + pd.Timedelta(days=1): event_date + pd.Timedelta(days=7)]
    
    pct_change = ((after_window[sector_names].mean() - before_window[sector_names].mean()) /
                  before_window[sector_names].mean()) * 100
    
    impact_matrix[event_name] = pct_change

# Step 4: Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(impact_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Sector Impact (% Change) Around Tariff Events", fontsize=14)
plt.xlabel("Tariff Event")
plt.ylabel("Sector")
plt.tight_layout()
plt.show()
```
